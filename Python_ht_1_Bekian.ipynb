```python
import pandas as pd
import numpy as np

# Реальные средние температуры (примерные данные) для городов по сезонам
seasonal_temperatures = {
    "New York": {"winter": 0, "spring": 10, "summer": 25, "autumn": 15},
    "London": {"winter": 5, "spring": 11, "summer": 18, "autumn": 12},
    "Paris": {"winter": 4, "spring": 12, "summer": 20, "autumn": 13},
    "Tokyo": {"winter": 6, "spring": 15, "summer": 27, "autumn": 18},
    "Moscow": {"winter": -10, "spring": 5, "summer": 18, "autumn": 8},
    "Sydney": {"winter": 12, "spring": 18, "summer": 25, "autumn": 20},
    "Berlin": {"winter": 0, "spring": 10, "summer": 20, "autumn": 11},
    "Beijing": {"winter": -2, "spring": 13, "summer": 27, "autumn": 16},
    "Rio de Janeiro": {"winter": 20, "spring": 25, "summer": 30, "autumn": 25},
    "Dubai": {"winter": 20, "spring": 30, "summer": 40, "autumn": 30},
    "Los Angeles": {"winter": 15, "spring": 18, "summer": 25, "autumn": 20},
    "Singapore": {"winter": 27, "spring": 28, "summer": 28, "autumn": 27},
    "Mumbai": {"winter": 25, "spring": 30, "summer": 35, "autumn": 30},
    "Cairo": {"winter": 15, "spring": 25, "summer": 35, "autumn": 25},
    "Mexico City": {"winter": 12, "spring": 18, "summer": 20, "autumn": 15},
}

# Сопоставление месяцев с сезонами
month_to_season = {12: "winter", 1: "winter", 2: "winter",
                   3: "spring", 4: "spring", 5: "spring",
                   6: "summer", 7: "summer", 8: "summer",
                   9: "autumn", 10: "autumn", 11: "autumn"}

# Генерация данных о температуре
def generate_realistic_temperature_data(cities, num_years=10):
    dates = pd.date_range(start="2010-01-01", periods=365 * num_years, freq="D")
    data = []

    for city in cities:
        for date in dates:
            season = month_to_season[date.month]
            mean_temp = seasonal_temperatures[city][season]
            # Добавляем случайное отклонение
            temperature = np.random.normal(loc=mean_temp, scale=5)
            data.append({"city": city, "timestamp": date, "temperature": temperature})

    df = pd.DataFrame(data)
    df['season'] = df['timestamp'].dt.month.map(lambda x: month_to_season[x])
    return df

# Генерация данных
data = generate_realistic_temperature_data(list(seasonal_temperatures.keys()))
data.to_csv('temperature_data.csv', index=False)
```

```python
import time


def get_moving_stats_and_anomalies(data, window=30):
    data['moving_avg'] = data['temperature'].rolling(window=window).mean()
    data['moving_std'] = data['temperature'].rolling(window=window).std()
    data['is_anomaly'] = (data['temperature'] > data['moving_avg'] + 2 * data['moving_std']) | \
                         (data['temperature'] < data['moving_avg'] - 2 * data['moving_std'])
    return data


def get_seasonal_stats(data):
    seasonal_stats = data.groupby(['city', 'season'])['temperature'].agg(['mean', 'std']).reset_index()
    return seasonal_stats


sequential_start = time.time()

sequential_data = get_moving_stats_and_anomalies(data)
seasonal_stats = get_seasonal_stats(data)

sequential_end = time.time()
```

```python
from concurrent.futures import ProcessPoolExecutor


parallel_start = time.time()

with ProcessPoolExecutor() as executor:
    future1 = executor.submit(get_moving_stats_and_anomalies, data)
    future2 = executor.submit(get_seasonal_stats, data)
    parallel_data = future1.result()
    seasonal_stats = future2.result()

parallel_end = time.time()
```

```python
import asyncio
import nest_asyncio
import aiohttp

nest_asyncio.apply()


async def get_city_weather(session, city, api_key):
    async with session.get(f"http://api.openweathermap.org/geo/1.0/direct?q={city}&limit=1&appid={api_key}") as response:
        response = await response.json()
        lat = response[0]['lat']
        lon = response[0]['lon']

    async with session.get(f"https://api.openweathermap.org/data/2.5/weather?lat={lat}&lon={lon}&exclude=minutely,hourly,daily,alerts&appid={api_key}&units=metric") as response:
        response = await response.json()
        return response['main']['temp']


async def main():
    current_temperature = dict()
    async with aiohttp.ClientSession() as session:
        tasks = [get_city_weather(session, city, API_KEY) for city in unique_cities]
        temperatures = await asyncio.gather(*tasks)

    for city, temp in zip(unique_cities, temperatures):
        current_temperature[city] = temp


parallel_start = time.time()

asyncio.run(main())

parallel_end = time.time()
```
